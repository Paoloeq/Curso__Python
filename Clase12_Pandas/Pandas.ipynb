# PANDAS - OPERACIONES

```python
#QUITAR A LA COLUMNA PRECIO "$" CON ""(VACIO) Y TRANSFORMARLO A UN FLOAT, ESTO SE HACE PARA HACER OPERACIONES
```

```python
preciosdf["Precio"]=preciosdf["Precio"].str.replace("$","").astype("float")
```

```python
preciosdf.head(5) #VER LA TABLA LOS PRIMEROS 5
```

```python
#QUITAR A LA COLUMNA COSTO "$" CON ""(VACIO) Y TRANSFORMARLO A UN FLOAT, ESTO SE HACE PARA HACER OPERACIONES
```

```python
preciosdf["Costo"]=preciosdf["Costo"].str.replace("$","").astype("float")
```

```python
preciosdf.head(4) # VISUALIZAR LOS PRIMEROS 4 EN LA TABLA
```

```python
#CREAR UNA NUEVA COLUMNA "GANANCIAS", Y EL VALOR SE DENOTA POR LA OPERACIÓN RESTA ENTRE "PRECIO" Y "COSTO"
```

```python
preciosdf["Ganancias"]=preciosdf["Precio"]-preciosdf["Costo"]
```

```python
preciosdf.head(5) 
```

```python
#BORRAR COLUMNA
```

```python
restaurantedf= restaurantedf.drop("Hora de Cobro",axis=1)
```

### Filter

```python
#CREAR UNA DATAFRAME "precioscarosdf" Y SOBRE ESTE SE VISUALIZA LOS PRECIOS > 100
```

```python
precioscarosdf=preciosdf[preciosdf["Precio"]>100]
```

```python
precioscarosdf.head(100)
```

```python
#CAMBIAR NOMBRE DE COLUMNA 
```

```python
restaurantdf = restaurantedf.rename(columns={"Atendió":"Mesero"})
```

```python
restaurantdf
```

# MERGE (tipo buscar v en excel)

```python
#PARA CADA TIPO DE PRODUCTO EN EL DATA FRAME "restaurantdf" ADICIONAR LA TABLA DE PRECIO Y COSTO DE OTRO DATA FRAME 
```

```python
restpreciosdf = restaurantdf.merge(preciosdf, on=["Producto"], how="left")
```

# TABLAS DINÁMICAS

```python
#CREAR TABALA DINAMICA INDEX (FILAS), VALUES (CONTENIDO), SE SUMA VALORES 
```

```python
dinamicdf=pd.pivot_table(restpreciosdf, values="Precio", index="Categoria", aggfunc=np.sum)
```

```python
dinamicdf
```
